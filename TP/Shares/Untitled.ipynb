```python
import requests 
import json 
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import time
```

```python
def get_bybit_bars(symbol, interval, startTime, endTime):
 
    url = "https://api.bybit.com/v5/market/kline"
 
    startTime = str(int(startTime.timestamp()))
    endTime   = str(int(endTime.timestamp()))
 
    req_params = {"symbol" : symbol, 'interval' : interval, 'start' : startTime, 'end' : endTime}
     
    res = requests.get(url, params = req_params).text    
    return res
        
    json_data  = json.loads(res)['result']    
    
    return json_data

    df = pd.DataFrame(json_data)
 
    if (len(df.index) == 0):
        return None
     
    df.index = [dt.datetime.fromtimestamp(x) for x in df.open_time]
 
    return df
```

```python
get_bybit_bars('ETHUSD', 1, dt.datetime(2023, 1, 1), dt.datetime(2024, 2, 1))
```
